## Colocando os documentos em um dataframe

```python
docs_df = pd.DataFrame(documentos)
docs_df
```

## Gerando os embeddings

```python
model = "models/embedding-001"

def embed_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]

docs_df["Embeddings"] = docs_df.apply(lambda row: embed_fn(row["Título"], row["Conteúdo"]), axis=1)
docs_df
```
